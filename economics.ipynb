```python
import pandas as pd
```

```python
housing_df = pd.read_csv("./Resources/State_time_series.csv")
housing_df
```

```python
clean_housing_df = housing_df[['Date','RegionName', 'MedianListingPricePerSqft_AllHomes', 'MedianRentalPricePerSqft_AllHomes', 
                               'PctOfHomesDecreasingInValues_AllHomes', 'PctOfHomesIncreasingInValues_AllHomes']]
clean_housing_df
```

```python
clean_housing_df = clean_housing_df.dropna()
clean_housing_df
```

```python
clean_housing_df = clean_housing_df.rename(columns={"RegionName": "State"})
```

```python
clean_housing_df = clean_housing_df.groupby(['Date', 'State'])
```

```python
clean_housing_df = clean_housing_df.agg({
    'MedianListingPricePerSqft_AllHomes': 'mean',
    'MedianRentalPricePerSqft_AllHomes': 'mean',
    'PctOfHomesDecreasingInValues_AllHomes': 'mean',
    'PctOfHomesIncreasingInValues_AllHomes': 'mean'
}).reset_index()
```

```python
clean_housing_df.to_csv("./db_files/housing.csv", index = False)
```
