```python
import pandas as pd
import numpy as np
```

Extract Source: Zillow Rental Price Data
https://www.kaggle.com/paultimothymooney/zillow-house-price-data

```python
# Path
df = "State_MedianRentalPrice_AllHomes.csv"
```

```python
# Read csv and store into Pandas DF
rental_df = pd.read_csv(df, encoding="ISO-8859-1", index_col=[0])
rental_df
```

Transform

```python
# Rename Region Name column to State
med_rental_df = rental_df.rename(columns={"RegionName": "State",})
med_rental_df  
```

```python
#Drop NAN values
clean_rental_df = med_rental_df.dropna(how='any')
clean_rental_df
```

```python
clean_rental_df = clean_rental_df[['State', 'Median Rental Price']]
clean_rental_df
```
